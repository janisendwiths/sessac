```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import matplotlib as plt
import numpy as np
import random
```

```python
import chromedriver_autoinstaller
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
# 파이썬이 제어하는 크롬 빈화면 뜨는 지 확인하기(창 닫지 말 것)
```

```python
df1 = pd.DataFrame(columns=['pageindex', 'title', 'date', 'agree'])
# 전체 청와대 청원 크롤링 데이터 프레임
```

```python
for n in range(1401,1501) : # 100 단위로 뽑아보기 (1,101), (101, 201), (201,301), ...

    browser.get("https://www1.president.go.kr/petitions/?c=0&only=2&page="+ str(n) +"&order=1" ) # 빈 크롬 화면 청원 페이지로 바뀌는 거 확인해보기

    soup = BeautifulSoup(browser.page_source, 'html.parser')
    
    print(n)
    
    for i in soup.select("#cont_view > div.cs_area > div > div > div.ct_petitions > div.ct_petitions_area.ct_txt_left > div.ct_list1 > div.board.text > div.b_list.category.b_list2 > div.bl_body > ul > li > div"):
        t = i.find("div", class_="bl_subject").text[3:]
        print(t.strip())
        d = i.find("div", class_="bl_date light").text[7:]
        print(d.strip())
        a = i.find("div", class_="bl_agree cs").text[5:]
        print(a.strip())
        page = n
        df1 = df1.append({'title' : t,'date' : d,'agree' : a, 'pageindex' : page},ignore_index=True)

    time.sleep(random.uniform(1,10))
```

```python
print(len(df1))
# 본인이 여태 뽑은 페이지 수 * 7 이 맞는지 확인 아니면 아래 코드로 빈 부분 확인
```
